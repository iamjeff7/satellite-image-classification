```python
# Feature extraction
labels = []
features = []
count = 0
for folder in os.listdir(train_path):
    print(folder)
    for image in os.listdir(os.path.join(train_path,folder)):
        filename = os.path.join(train_path,folder,image)
        img = cv2.imread(filename)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        min_des = 100
        threshold = 0.05
        sift = cv2.xfeatures2d.SIFT_create(contrastThreshold=threshold)
        kps, des = sift.detectAndCompute(gray, None)
        try:
            l = len(des)
        except:
            threshold = 0.01
            sift = cv2.xfeatures2d.SIFT_create(contrastThreshold=threshold)
            kps, des = sift.detectAndCompute(gray, None)
            l = len(des)
        while (min_des>l):
            #print(threshold,l)
            threshold -= 0.001
            sift = cv2.xfeatures2d.SIFT_create(contrastThreshold=threshold)
            kps, des = sift.detectAndCompute(gray, None)
            l = len(des)
        
        labels.append(label.index(folder))
        features.append((kps,des))
        print('\t',image)
        
            
print(len(features))
print(len(features[0]))
#print(features[0][1])
```

```python
def extract_features(image):
    img = cv2.imread(image)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    min_des = 100
    threshold = 0.05
    sift = cv2.xfeatures2d.SIFT_create(contrastThreshold=threshold)
    kps, des = sift.detectAndCompute(gray, None)
    try:
        l = len(des)
    except:
        threshold = 0.01
        sift = cv2.xfeatures2d.SIFT_create(contrastThreshold=threshold)
        kps, des = sift.detectAndCompute(gray, None)
        l = len(des)
    while (min_des>l):
        #print(threshold,l)
        threshold -= 0.001
        sift = cv2.xfeatures2d.SIFT_create(contrastThreshold=threshold)
        kps, des = sift.detectAndCompute(gray, None)
        l = len(des)
    return des

def build_histogram(des):
    k = 50
    code, distortion = vq(des, codebook)
    bowhist,bin_edge = np.histogram(code, k, density=True)
    return bowhist
```

```python
from sklearn.neighbors import NearestNeighbors

filename = os.path.join(test_path,'beach','beach_551.jpg')
descriptor = extract_features(filename)
histogram = build_histogram(descriptor)
neighbor = NearestNeighbors(n_neighbors = 20)
neighbor.fit(bow)

dist, result = neighbor.kneighbors([histogram])
```

```python
print("Index of beach",label.index('beach'))
for j,i in zip(dist[0],result[0]):
    print(label[labels[i]],'\t\t',j)
```

```python
from sklearn.neighbors import NearestNeighbors

y_pred = []
y_true = []
filenames = []
neighbor = NearestNeighbors(n_neighbors = 20)
neighbor.fit(bow)
for folder in os.listdir(test_path):
    print(folder)
    for image in os.listdir(os.path.join(test_path,folder)):
        filename = os.path.join(test_path,folder,image)
        descriptors = extract_features(filename)
        histogram = build_histogram(descriptors)
        dist, result = neighbor.kneighbors([histogram])
        result = result[0]
        result = [labels[r] for r in result]
        predicted_label = np.argmax(np.bincount(result))
        y_pred.append(predicted_label)
        y_true.append(label.index(folder))
        filenames.append(filename)
```

```python
for i,j in zip(y_true, y_pred):
    print("{}\t{}".format(i,j))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
correct = 0
wrong = 0
wrong_image = []
for i,j,k in zip(y_true, y_pred, filenames):
    if i==j:
        correct += 1
    else:
        wrong += 1
        wrong_image.append([k,j])
print("Correct: {}".format(correct))
print("Wrong: {}".format(wrong))
print("Accuracy: {}%".format((correct/(correct+wrong))*100))

precision = precision_score(y_true, y_pred, labels=None, pos_label=1, average='macro', sample_weight=None)
recall = recall_score(y_true, y_pred, labels=None, pos_label=1, average='macro', sample_weight=None)
f1 = f1_score(y_true, y_pred, labels=None, pos_label=1, average='macro', sample_weight=None)
print("Precision: {}".format(precision))
print("Recall:    {}".format(recall))
print("F1:        {}".format(f1))
```
