```python

vgg_conv = VGG16(weights='imagenet',
                 include_top=False,
                input_shape=(224,224,3))

```

```python
train_dir = 'NWPU-RESISC12/train'
validation_dir = 'NWPU-RESISC12/test'
image_size = 224
nTrain = 6600
nTest = 1800
```

```python
from keras.preprocessing.image import ImageDataGenerator
# Train :: Extract Features
datagen = ImageDataGenerator(rescale=1./255)
batch_size = 20

train_features = np.zeros(shape=(nTrain, 7, 7, 512))
train_labels = np.zeros(shape=(nTrain,12))

train_generator = datagen.flow_from_directory(
    train_dir,
    target_size=(image_size, image_size),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=True)

i = 0
for inputs_batch, labels_batch in train_generator:
    features_batch = vgg_conv.predict(inputs_batch)
    train_features[i*batch_size:(i+1)*batch_size]=features_batch
    train_labels[i*batch_size:(i+1)*batch_size]=labels_batch
    i+=1
    print(i,end=' ')
    if i*batch_size >= nTrain:
        break
print('\Done')
train_features = np.reshape(train_features, (nTrain, 7 * 7 * 512))
```

```python
import pickle
with open('VGG_Bottleneck_Train_Features.pkl','wb') as handle:
    pickle.dump(train_features, handle, protocol=pickle.HIGHEST_PROTOCOL)
with open('VGG_Bottleneck_Train_Labels.pkl','wb') as handle:
    pickle.dump(train_labels, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
test_features = np.zeros(shape=(nTest, 7, 7, 512))
test_labels = np.zeros(shape=(nTest,12))

test_generator = datagen.flow_from_directory(
    test_dir,
    target_size=(image_size, image_size),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False)

i = 0
for inputs_batch, labels_batch in test_generator:
    features_batch = vgg_conv.predict(inputs_batch)
    test_features[i * batch_size : (i + 1) * batch_size] = features_batch
    test_labels[i * batch_size : (i + 1) * batch_size] = labels_batch
    i += 1
    if i * batch_size >= nTest:
        break

test_features = np.reshape(test_features, (nTest, 7 * 7 * 512))
```

```python
print('train:',train_features.shape)
print('test:',test_features.shape)

print('train_label:',train_labels.shape)
print('test_label:',test_labels.shape)
```

```python
y_true = []
for i in train_labels:
    y_true.append(np.argmax(i))
y_true
```

```python
# SVM
from sklearn import svm
clf = svm.SVC(gamma='scale', decision_function_shape='ovo')
clf.fit(train_features, y_true)
```

```python
clf.predict([test_features[0]])
```

```python
test_true = []
for i in test_labels:
    test_true.append(np.argmax(i))
```

```python
len(test_true)
```

```python
test_pred = []
for i in range(len(test_true)):
    test_pred.append(clf.predict([test_features[i]])[0])
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix

accuracy = accuracy_score(test_true,test_pred)
precision = precision_score(test_true,test_pred,average='micro')
recall = recall_score(test_true,test_pred,average='micro')
f1 = f1_score(test_true,test_pred,average='micro')
print("Accuracy:  {}".format(accuracy))
print("Precision: {}".format(precision))
print("Recall:    {}".format(recall))
print("F1:        {}".format(f1))
```

```python
cm = confusion_matrix(test_true,test_pred)
print(cm)
```
